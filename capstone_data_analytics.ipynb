## Average Customer Order

Average Customer Order is a key performance indicator (KPI) that measures the average number of products purchased by each customer. It is calculated by dividing the total number of products sold by the total number of customers.

This metric is important for businesses as it provides insights into customer buying behavior. A high Average Customer Order might indicate that customers are finding many products they like, or that promotional strategies like bundling are working effectively. On the other hand, a low Average Customer Order might suggest that customers are not finding enough products they like, or that they are only buying the minimum amount necessary.

By monitoring the Average Customer Order, businesses can gain valuable insights to help them improve their product offerings, marketing strategies, and overall customer satisfaction. For example, if the Average Customer Order is lower than expected, a business might decide to introduce new products, offer bundle deals, or implement a loyalty program to encourage customers to buy more.

```python
# Define 'customer_id' and 'quantity_sales' column names
customer_id_column = 'customer_id'
quantity_sales_column = 'quantity_sales'

# Convert 'quantity_sales' column to numeric data type
df[quantity_sales_column] = pd.to_numeric(df[quantity_sales_column], errors='coerce')

# Calculate total quantity for each customer
total_quantity_per_customer = df.groupby(customer_id_column)[quantity_sales_column].sum()

# Calculate average quantity per customer
average_quantity_per_customer = total_quantity_per_customer.mean()

print("Average quantity per customer: ", average_quantity_per_customer)

# Get top 20 customers with highest total quantity
top_20_customers = total_quantity_per_customer.nlargest(20)

# Visualization
top_20_customers.plot(kind='bar', figsize=(10,6))
plt.title('Total Quantity per Customer (Top 20)')
plt.xlabel('Customer ID')
plt.ylabel('Total Quantity')
plt.show()
```

## Visualization of Average Customer Order

The bar chart represents the total quantity of products ordered by the top 20 customers. Each bar in the chart represents a unique customer, identified by their customer ID on the x-axis. The height of the bar represents the total quantity of products that the customer has ordered, as shown on the y-axis.

From the chart, we can observe which customers have ordered the most products. This can help us identify our most valuable customers, who contribute a significant portion of our sales. These customers might be targeted with special offers or loyalty programs to encourage them to continue buying our products.

Moreover, by comparing the heights of the bars, we can also get a sense of the variation in order quantities among our top customers. This can provide insights into the range of our customers' buying behaviors, which can be useful for tailoring our marketing strategies to different customer segments.

## Average Shipping Time

Average Shipping Time is a key performance indicator (KPI) that measures the average time it takes for products to be delivered to customers after they have been ordered. It is calculated by subtracting the order date from the delivery date for each order, and then taking the average of these values.

This metric is crucial for businesses as it directly impacts customer satisfaction. Customers expect their orders to be delivered within a reasonable time frame, and any delays can lead to dissatisfaction and negative reviews. By monitoring the Average Shipping Time, businesses can identify any issues in their delivery process and take steps to address them, thereby improving customer satisfaction and loyalty.

Moreover, understanding the Average Shipping Time can also help businesses in planning their inventory. If the shipping time is long, businesses might need to keep more stock to ensure they can meet customer demand. On the other hand, if the shipping time is short, they might be able to operate with less stock, thereby reducing inventory costs.

```python
# Define 'order_date' and 'delivery_date' column names
order_date_column = 'order_date' 
delivery_date_column = 'delivery_date' 

# Convert 'order_date' and 'delivery_date' to datetime
df[order_date_column] = pd.to_datetime(df[order_date_column])
df[delivery_date_column] = pd.to_datetime(df[delivery_date_column])

# Calculate shipping time for each order
df['shipping_time'] = (df[delivery_date_column] - df[order_date_column]).dt.days

# Calculate average shipping time
average_shipping_time = df['shipping_time'].mean()

print("Average shipping time: ", average_shipping_time)

# Visualization
df['shipping_time'].plot(kind='hist', rwidth=0.8)
plt.title('Distribution of Shipping Time')
plt.xlabel('Shipping Time (days)')
plt.show()
```

## Visualization of Average Shipping Time

The histogram represents the distribution of shipping times for all orders. Each bar in the histogram represents the number of orders (frequency) that had a shipping time within a particular range of days.

The x-axis represents the shipping time in days, and the y-axis represents the frequency of orders. For example, a bar reaching up to 50 on the y-axis for the range of 5-6 days on the x-axis means that there were 50 orders that were delivered in 5-6 days.

The curve on the histogram (known as the Kernel Density Estimate or KDE) gives a smooth outline of the distribution, making it easier to understand the overall pattern.

From the histogram, we can observe the most common shipping times, as well as the range and distribution of shipping times. This can help us understand how consistently the company is able to deliver orders within certain time frames, which is crucial for maintaining customer satisfaction.

## Average Sales per Product

Average Sales per Product is a key performance indicator (KPI) that measures the average number of units sold for each product. It is calculated by dividing the total number of units sold by the total number of unique products.

This metric is important for businesses as it provides insights into product performance. A high Average Sales per Product might indicate that the product range is well-curated and each product is contributing significantly to sales. On the other hand, a low Average Sales per Product might suggest that some products are not selling well and may need to be re-evaluated or discontinued.

By monitoring the Average Sales per Product, businesses can identify which products are underperforming and take steps to improve their sales, such as through targeted marketing campaigns, pricing adjustments, or product improvements. This can help businesses optimize their product range and increase overall sales.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Final Data Cleaning - cleaning.csv')

# Convert 'order_date' to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Set 'order_date' as the index
df.set_index('order_date', inplace=True)

# Sort the index in ascending order
df.sort_index(inplace=True)

# Filter the data for the specified 10 months
df_2021 = df.loc['2021-01-01':'2021-10-31']

# Calculate total quantity for each product
total_quantity_per_product = df_2021.groupby('product_id')['quantity_sales'].sum()

# Calculate average sales per product in the last 10 months
average_sales_per_product = total_quantity_per_product / 10
print("Average sales per product from January to October 2021: ", average_sales_per_product)

# Calculate average sales per month
average_sales_per_month = df_2021.resample('M')['quantity_sales'].sum() / df_2021['product_id'].nunique()

# Plot average sales per month
average_sales_per_month.plot(kind='line')
plt.title('Average Sales per Product per Month (Jan-Oct 2021)')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.show()
```

## Visualization of Average Sales per Product

The bar chart represents the average sales for each product. Each bar in the chart represents a unique product, identified by their product ID on the x-axis. The height of the bar represents the average number of units sold for that product, as shown on the y-axis.

From the chart, we can observe which products have the highest and lowest average sales. This can help us identify our best-selling and worst-selling products. The best-selling products might be promoted more heavily, while the worst-selling products might need to be re-evaluated or discontinued.

Moreover, by comparing the heights of the bars, we can also get a sense of the variation in sales among our products. This can provide insights into the diversity of our product performance, which can be useful for making strategic decisions about product development and marketing.

## Report which product have the highest sales in the last 10 months (Jan - Oct 2021)

This report aims to identify the product that has achieved the highest sales in the last 10 months, from January to October 2021.

```python
# Group by product_id and count the number of unique order_id for each product
product_orders = df_2021.groupby(['product_id', 'product_name'])['order_id'].nunique()

# Group by product_id and sum the quantity for each product
product_quantity = df_2021.groupby(['product_id', 'product_name'])['quantity_sales'].sum()

# Combine the two series into a DataFrame
product_info = pd.DataFrame({'orders': product_orders, 'quantity_sales': product_quantity})

# Calculate a new column 'sales' as the product of 'orders' and 'quantity'
product_info['sales'] = product_info['orders'] * product_info['quantity_sales']

# Find the top 10 products with the highest sales
top_products = product_info['sales'].nlargest(10)

# Plot the data
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=top_products.values, y=top_products.index.get_level_values(1))

# Add labels to the plot
for i, (value, name) in enumerate(zip(top_products.values, top_products.index.get_level_values(1))):
    barplot.text(value, i, f'{value:,.0f}', ha = 'left')

plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.title('Top 10 Products with the Highest Sales from Jan - Oct 2021')
plt.show()
```

Highest Selling Product (Jan - Oct 2021)

After analyzing the sales data from January to October 2021, we found that the product with the highest sales is **Flannel**.

This information can be used to prioritize inventory management and marketing efforts for these high-selling products.

### Identify trends in the market and forecast future sales

In this capstone project, we will use historical sales data to identify patterns and trends in the market. The ultimate goal is to forecast future sales of our specific product line. This process is crucial in our business context for several reasons:

1. **Planning**: By understanding sales trends, we can better plan our production and inventory. This will help us to meet customer demand without overproducing.
2. **Strategic Decision Making**: Sales forecasts can assist us in making strategic decisions. For example, they can inform us when to launch new products, when to ramp up production, and when to discontinue certain products.
3. **Identifying Opportunities**: Analyzing market trends can help us identify new opportunities. This could be untapped markets where our products could be successful, or gaps in the market that could be filled with new product development.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('Final Data Cleaning - cleaning.csv')

# Ensure the date column is in datetime format and set as index
df['order_date'] = pd.to_datetime(df['order_date'])
df = df.set_index('order_date')

# Calculate monthly sales
monthly_sales = df.resample('M')['quantity_sales'].sum()

# Calculate moving average
moving_avg = monthly_sales.rolling(window=12).mean()

# Plot original time series and moving average
plt.plot(monthly_sales, label='Original')
plt.plot(moving_avg, color='red', label='Moving Average')
plt.legend(loc='best')
plt.title('Identify Market Trends')
plt.show()
```


After analyzing the monthly sales data, we identified several key trends in the market:

1. **Seasonality**: Sales tend to peak in the fourth quarter of the year, which could be attributed to the holiday season.

2. **Growth**: There is a clear upward trend in sales, indicating growth in the market.

3. **Moving Average**: The 10-month moving average shows a steady increase in sales over time, further confirming the growth trend.

These trends provide valuable insights for future business strategies. For instance, promotional activities could be timed to coincide with the peak sales period in the fourth quarter. The overall growth trend also suggests potential for expansion.
